```python
import pandas as pd

# Upload file in Colab
from google.colab import files
uploaded = files.upload()

# Load dataset
df = pd.read_excel(list(uploaded.keys())[0])

# Preview
df.head()
```

```python
print(df.columns.tolist())
```

```python
# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(".", "").str.replace(" ", "_")

print(df.columns)
```

```python
df.head()
```

```python
import re

def convert_employees(x):
    if pd.isna(x):
        return 0

    x = str(x).lower()

    # Handle "lakh"
    if "lakh" in x:
        num = re.findall(r'\d+', x)
        if num:
            return int(num[0]) * 100000

    # Handle normal numbers
    num = re.findall(r'\d+', x)
    if num:
        return int(num[0])

    return 0

df['employees_clean'] = df['employees'].apply(convert_employees)
```

```python
emp_data = df.groupby('name')['employees_clean'].sum().head(10)

import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))
plt.pie(emp_data, labels=emp_data.index, autopct='%1.1f%%')
plt.title("Employees Distribution")
plt.show()
```

```python
import pandas as pd
import re

# Clean ratings (if text)
df['ratings'] = pd.to_numeric(df['ratings'], errors='coerce')

# Clean reviews (remove commas etc.)
df['review_count'] = df['review_count'].astype(str).str.replace(",", "")
df['review_count'] = pd.to_numeric(df['review_count'], errors='coerce')

# Clean years (if needed)
df['years'] = pd.to_numeric(df['years'], errors='coerce')
```

```python
print(df.columns.tolist())
```

```python
df['review_count'] = (
    df['review_count']
    .astype(str)
    .str.replace(",", "")
    .str.extract('(\d+)')[0]
)

df['review_count'] = pd.to_numeric(df['review_count'], errors='coerce')
```

```python
import plotly.express as px

review_data = df.groupby('name')['review_count'].sum().dropna().sort_values(ascending=False).head(10)

print(review_data)  # 🔥 must show values

fig = px.funnel(
    x=review_data.values,
    y=review_data.index,
    title="Top Companies by Reviews"
)

fig.show()
```

```python
import matplotlib.pyplot as plt

rating_data = df.groupby('name')['ratings'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(rating_data.index, rating_data.values)
plt.xticks(rotation=45)
plt.title("Top Companies by Rating")
plt.ylabel("Rating")
plt.show()
```

```python
df['years'] = pd.to_numeric(df['years'], errors='coerce')

year_data = df.groupby('years')['employees_clean'].sum().dropna()

print(year_data)  # 🔥 must NOT be empty
```
